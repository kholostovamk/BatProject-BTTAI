Site.Access (6) ; Site.Classification (2); Site.Ownership (1) since data is categorical it makes sense to impute these columns. Want to use .mode() which puts the value that appears most often.


```python
columns_to_impute = ['Site.Access', 'Site.Classification', 'Site.Ownership']

for col in columns_to_impute:
  mode_value = df[col].mode()[0] #0 in the end needed to take the first value out of .mode(), because sometimes there can be multiple
  df[col].fillna(mode_value,inplace=True)
  print("This is mode value for " + col + " " + mode_value)
```

```python
# Checking the number of missing values after imputation
missing_values_after_imputation = df[columns_to_impute].isnull().sum()
missing_values_after_imputation
```

Not sure what technique for outliers would be better
---

    Capping:
        Pros: Preserves most of the data and only adjusts extreme values. Reduces the influence of extreme outliers.
        Cons: Arbitrarily alters the genuine extreme values, which might not be ideal if those values have significance.

    Transformation:
        Pros: Makes the data more normally distributed and reduces the influence of outliers. It can help if subsequent modeling techniques assume normality.
        Cons: Alters the scale and interpretation of the data. This might be problematic if direct counts and their original scale are important for interpretation.

    Removal:
        Pros: Eliminates extreme values that might be errors or anomalies.
        Cons: Reduces the dataset size and might remove genuine extreme observations that could be significant.

```python
from matplotlib.lines import Line2D

# Count the occurrences of WNS for each state or province
state_counts = merged_df[merged_df['label'].isin(['Confirmed - Pos', 'White-nose Syndrome', 'WNS Positive'])]['state'].value_counts()

# Select the top 15 most affected states/provinces
top_15_counts = state_counts.head(15)

# Map back the state abbreviations to their full names for the top 15
state_abbreviation_mapping = {v: k for k, v in state_name_mapping.items()}
top_15_full_names = [state_abbreviation_mapping.get(state, state) for state in top_15_counts.index]

# Determine the colors for each state in the top 15 based on the country information in the DataFrame
state_country_mapping = dict(zip(merged_df['state'], merged_df['country']))
colors_final = ['blue' if state_country_mapping[state] == 'USA' else 'red' for state in top_15_counts.index]

# Plot the counts for the top 15 most affected states/provinces with full names and corrected colors
plt.figure(figsize=(10, 6))
ax = top_15_counts.plot(kind='bar', color=colors_final)
plt.title('Top 15 States and Provinces Most Affected by WNS')
plt.xlabel('State/Province')
plt.ylabel('Number of Occurrences')
plt.xticks(ticks=range(len(top_15_full_names)), labels=top_15_full_names, rotation=45)
plt.tight_layout()

# Add a legend for clarity
legend_elements = [Line2D([0], [0], color='blue', lw=4, label='USA'), Line2D([0], [0], color='red', lw=4, label='Canada')]
ax.legend(handles=legend_elements, loc='upper right')

plt.show()
```

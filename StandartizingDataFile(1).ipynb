FROM METADATA

**EQUIV** =  Non-negative results by qPCR2 that are outside the range of accepted, standardized laboratory-defined criteria for Pd Positive.  **WILL BE PD NOT DETECTED**

**NEG** = Characteristic histologic lesions of WNS are not present AND bat is negative for Pd (either by qPCR2 or fungal culture)

**SUSPECT** = Suspect for WNS – (one of the following criteria must be met)
a) Characteristic histologic lesions of WNS1 are present on an individual bat but Pd is not detected, the test result for Pd is inconclusive (either by qPCR2 or by fungal culture), or further testing for Pd is not performed.
b) One or more field signs* are observed in a bat colony AND Pd is detected (either by qPCR2, fungal culture, or a tapelift performed directly on visible fungal growth on bat skin) on at least one individual of the same species AND histopathology is negative or is not performed.
c) MULTIPLE field signs* are observed among species known to be susceptible to WNS and are within the currently recognized range of WNS but no samples are collected for diagnostic evaluation.
o	d) Individual bats that are part of a confirmed WNS morbidity/mortality event are submitted to, but not tested by, a diagnostician. This criterion is for instances in which multiple samples from the same site are submitted, but only a subset of those samples is tested. The untested samples may be classified as suspect for WNS if the subset of tested samples is Positive for WNS and consists of the same species as the untested samples. Representatives of all species involved in the disease event should be tested  

**Confirmed - Pos** - Characteristic histologic lesions of WNS are present on an individual bat AND the bat is positive for Pseudogymnoascus destructans (Pd) either by qPCR2 or by fungal culture.
        
**Pd not detected** - Pd is not detected (either by qPCR or fungal culture)

**Pd present** - Pd detected by qPCR or by fungal culture

```python
#list for having names of columns that will be renamed to WNS
wns_label = [label_names[i] for i in [6,7,9,10,11,12,13,14,15,17]]
wns_label
```

```python
#list for having names of columns that will be renamed to PD
pd_label = [label_names[i] for i in [0,2,16,19]]
pd_label
```

```python
#list for having names of columns that will be renamed to NEG (negative)

neg_label = [label_names[i] for i in [4,5,18,3,8,1]]
neg_label
```

```python
for label in neg_label:
    df.loc[df['label'] == label, 'label'] = 'NEG'

for pd_label in pd_label:
    df.loc[df['label'] == pd_label, 'label'] = 'PD'

for wns_label in wns_label:
    df.loc[df['label'] == wns_label, 'label'] = 'WNS'
```

Aside from label columns there are columns Total.Live,	Total.Dead, Total.Bat.Count that need to be standartized. Some of them has numbers, some says 'no data', and some has empty cells. No Data and empty collumns will contain zeros instead

```python
columns_to_modify = ['Total.Live', 'Total.Dead', 'Total.Bat.Count']

for col in columns_to_modify:
    df.loc[df[col] == 'No Data', col] = 0  # Replace 'No Data' with 0
    df[col].fillna(0, inplace=True)        # Replace NaN or empty values with 0
```
